```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r"C:\Users\User\Desktop\data_02.csv")
```

```python
data.head()
```

```python
class_ids={'Dobermann':0, 'German Shepherd':1, 'Rottweiler':2}
data['Class ID']=data['Class'].apply(lambda x: class_ids[x])
```

```python
data.head()
```

```python
fig, ax=plt.subplots()
ax.scatter(x=data['Weight'], y=data['Height'], c=data['Class ID'])
```

```python
model=SVC(kernel='linear')
```

```python
model
```

```python
model.fit(data[['Weight','Height']], data['Class ID'])
```

```python
X_test=np.random.rand(10000,2)
X_test=X_test*(40,20)+(25,55)
```

```python
y_pred=model.predict(X_test)
```

```python
fig, ax=plt.subplots()
ax.scatter(x=X_test[:,0], y=X_test[:,1], c=y_pred, alpha=.25)
```

```python
fig, ax=plt.subplots()
ax.scatter(x=X_test[:,0], y=X_test[:,1], c=y_pred, alpha=.25)
colors=['b','r','g']
y_colors=data['Class ID'].apply(lambda x:colors[x])
ax.scatter(x=data['Weight'], y=data['Height'], c=y_colors)
```
